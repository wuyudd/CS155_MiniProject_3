```python
def load_shakespeare(filename):
    peoms = []
    count = 0
    with open(filename, 'r') as f:
        peom = ""
        for line in f.readlines():
            if line[-1] != "\n": # the last peom
                peoms.append(peom)
            line = line.strip()
            if line.isdigit(): # digit for different peoms
                continue
            elif len(line) == 0: # 2 empty line, this peom ends
                count += 1
                if count == 2:
                    peoms.append(peom.strip())
                    peom = ""
                    count = 0
            else:
                line += " "
                peom += line
    return peoms
```

```python
shakespeare_peoms = load_shakespeare("./project3/data/shakespeare.txt")
print("# of peoms = ", len(shakespeare_peoms))
print(shakespeare_peoms)
# for peom in shakespeare_peoms:
#     print(peom)
#     print("**************************")
```

```python
def load_shakespeare_sentences(filename):
    sentences = []
    count = 0
    with open(filename, 'r') as f:
        for line in f.readlines():
            line = line.strip()
            if len(line) == 0: # empty line
                continue
            elif line.isdigit(): # peom number
                #print("# of peom = ", int(line)-1, ", count = ", count)
                count = 0
                continue
            else: # sentence
                sentences.append(line)
                count += 1
    return sentences 
```

```python
shakespeare_sentences = load_shakespeare_sentences("./project3/data/shakespeare.txt")
print("number of sentences = ", len(shakespeare_sentences))
for sentence in shakespeare_sentences:
    print(sentence)
```
